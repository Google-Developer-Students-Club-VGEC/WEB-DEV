```python
1. How and where can we insert the javascript in our html file?

-> We can add Javascript inside an HTML file using tag <script>.
-> The <script> tag can be placed in the <head> section of your HTML or in the <body> section, 
   depending on when you want the JavaScript to load.
-> Generally, js code goes inside <head> in order to keep them contained and out of the main content.
-> We can use inline method when we have to call a function in the HTML event attributes. 
   Like onClick, where we add code directly inside an html attribute.
-> We can create a separate file to write js code with .js extension. 
   And include it in our HTML document using src attribute.
-> It is useful when you have to write same code in multiple files.
    
For ex.  
    <html>  
        <head>  
            <title>Title</title>  
            <script>  
                document.write("Hello World!");  
            </script>  
        </head>  
        <body>  
        	<p>Hello World!</p>  
        </body>  
    </html>  
    
    
```

```python
2. Why we inspect tab (i.e. when we open a webpage we can inspect it with: right click->Inspect)

-> We can inspect the page to view our html code and view console where we can see our js output. 
   For ex. using console.log(), we can see the output in console tab in inspect.
-> We can also change the html/css of the page to see the changes temporarily. 
   It doesnt affect the original document. Once you reload the page it goes away.
```

```python
3. document.getElementById()

-> This method returns an element with the specified value.
-> It is used every time you want to read or edit an HTML element.
Syntax:
    document.getElementById(elementID)

For ex.
    <html>
    <body>
    <h1>The Document Object</h1>
    <h2>The getElementById() Method</h2>
    <p id="demo"></p>

    <script>
        document.getElementById("demo").innerHTML = "Hello World";
    </script>

    </body>
    </html>
```

```python
4. document.write()

-> This method writes directly to an HTML document. It works like printf() of c language.
-> Syntax:
    document.write(exp1, exp2, ..., expN)
    
For ex.
    <html>
    <body>

    <h2>The write() Method</h2>

    <script>
    document.write("Hello World!");
    </script>

    </body>
    </html>
```

```python
5. console.log()

-> The console.log() method outputs a message to the web console.
-> You can see the console in inspect tab.
-> It is generally used for testing purpose.

Syntax:
    console.log(message)

For ex:
    <html>
    <body>

    <h2>The log() Method</h2>

    <script>
        console.log("Hello world!");
    </script>

    </body>
    </html>
```

```python
6. <script> tag and values we can pass in script tag (like src, type)

-> The <script> tag is used to embed a client-side script i.e. JavaScript.
-> It contains either the javascript code or points to external file using src attribute.

src - Specifies the URL of an external script file
type - Specifies the media type of the script
```

```python
7. Basic Input/Output

-> We can take input using prompt() function.

prompt() : We use the prompt() function to ask the user for input. 
    As a parameter, we input the text we want to display to the user. 
    Once the user presses “ok,” the input value is returned.
    syntax - var var_name = prompt("text");
    
    for ex: 
        var name = prompt("What is your name?");
        var num = prompt("What is your roll number? ");

        println("Hello " + name + "!");
        println(num + "is your roll number, right?");

-> We can show output using innerHTML, document.write() and window.alert(). 
-> We can use console.log() to show output for debugging and testing purposes.

innerHTML : The innerHTML property sets or returns the HTML content of an element.
    It is commonly used after the document.getElementById().
    syntax - element.innerHTML = text
    
    for ex:
        <html>
        <body>
        
            <h2>The innerHTML Property</h2>

            <p id="demo"></p>

            <script>
                document.getElementById("demo").innerHTML = "Hello World!";
            </script>

        </body>
        </html>
        
document.write() : The write() method is used to write some content or JavaScript code in a Document.
    It works like printf() of c language.
    syntax - document.write(exp1, exp2, ..., expN)
    
    for ex:
        <html>
        <body>

            <h2>The write() Method</h2>

            <script>
            document.write("Hello World!");
            </script>

        </body>
        </html>
        
window.alert() : This method displays an alert box with a message and an OK button.
    The alert box takes the focus away from the current window, and forces the user to read the message.
    syntax - window.alert()
    
    for ex:
        <html>
        <body>

            <h2>The alert() Method</h2>	

            <script>
                window.alert("Hello World!");
            </script>

        </body>
        </html>
```

```python
8. Var

-> We can declare variables in js using 3 ways: var, let, const 

var: The variables that are defined with var statement have function scope.
    We can declare a variable again even if it has been defined previously in the same scope using var.
    Hoisting is allowed with var. (Hoisting means a function or a variable can be used before declaration.)
    
    syntax:
        var name = value;
        
    For ex:
        <html>
        <body>

        <h1>var example</h1>

        <p id="demo"></p>

        <script>
            var x = 5;
            var y = 6;
            var z = x + y;
            document.getElementById("demo").innerHTML ="Sum is: " + z;
        </script>

        </body>
        </html>
        
let: The variables that are defined with let statement have block scope.
    We cannot declare a variable more than once if we defined that previously in the same scope.
    Hoisting is not allowed with let.
    
    syntax:
        let name = value;
        
    For ex:
        <html>
        <body>

        <h2>let Example</h2>

        <p id="demo"></p>

        <script>
            let x = 5;
            let y = 6;
            document.getElementById("demo").innerHTML ="Sum is: " + (x + y);
        </script>

        </body>
        </html>
        
const: The values defined with const cannot be changed.
    
    syntax:
        const name = value;
        
    For ex:
        <html>
        <body>

        <h1>const Example</h1>

        <p id="demo"></p>

        <script>
            const a = 5;
            const b = 6;
            let c = a + b;
            document.getElementById("demo").innerHTML = "Sum is: " + c;
        </script>

        </body>
        </html>
```
